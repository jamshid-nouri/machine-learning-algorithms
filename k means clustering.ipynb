```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X =iris.data
y= iris.target
X[0:5]
```

```python
import matplotlib.pyplot as plt
plt.subplot(121)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel("X0")
plt.ylabel("X1")

plt.subplot(122)
plt.scatter(X[:,2],X[:,3],c=y)
plt.xlabel("X2")
plt.ylabel("X3")
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=10)
```

```python
kmeans.fit(X)
kmeans.labels_
kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt
plt.subplot(221)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel("X0")
plt.ylabel("X1")
plt.title('true y')

plt.subplot(222)
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.xlabel("X0")
plt.ylabel("X1")
plt.title('kmeans')

plt.subplot(223)
plt.scatter(X[:,2],X[:,3],c=y)
plt.xlabel("X2")
plt.ylabel("X3")
plt.title('true y')

plt.subplot(224)
plt.scatter(X[:,2],X[:,3],c=kmeans.labels_)
plt.xlabel("X2")
plt.ylabel("X3")
plt.title('kmeans')
```
