```python
 import numpy as np
```

```python
def gradient_destent(x,y):
    w = 0
    b = 0
    iterations = 100
    m = len(x)
    learning_rate = 0.1
    
    for i in range(iterations):
        y_predict = w * x + b
        J = 1/(2*m)*np.sum(np.dot(y_predict-y,y_predict-y))
        w = w - learning_rate * 1 / m * np.sum(np.dot(y_predict-y,x))
        b = b - learning_rate * 1 / m * np.sum(y_predict-y)
        print('w = ',w , 'b = ' ,b ,'J = ',J )
    return w,b,J
        
```

```python
x = np.array([1,2,3])
y = np.array([1,2,3])
```

```python
gradient_destent(x,y)
```
